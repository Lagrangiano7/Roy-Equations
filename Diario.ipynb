## Objetivo 1: consecuencia sobre $T$ de la unitariedad de $S$

- En un proceso de scattering de dos partículas, estas pueden interaccionar o no. Si no interaccionan, el estado de salida ha de ser el mismo que el de entrada, mientras que si sí interaccionan, esta interacción ha de conservar el cuadrimomento del sistema
- De este modo, es natural separar el operador de scattering, que dado un ket de entrada nos dice cuál será el de salida, como:

$$S=I+iT$$

- Naturalmente, si tomamos elementos de matriz entre dos estados entrada-salida (i-f) concretos, entonces el operador ha de reflejar la condición que hemos impuesto:

$$\langle f | S |i\rangle = \delta_{fi}+i(2\pi)^4\delta^4(p_f-p_i)\langle f |T|i\rangle$$

- Volvamos ahora a la expresión de operadores. Estudiamos qué consecuencias tiene sobre $T$ el hecho de que $S$ sea un operador unitario:

$$SS^\dagger=I \Longrightarrow (I+iT)(I-iT^\dagger)=I$$

- Desarrollando, llegamos a la expresión en operadores para $T$:

$$TT^\dagger =i(T^\dagger-T)$$

- Tomando ahora elementos de matriz para estados inicial-final (lo que me interesa es ver qué condición van a cumplir estos elementos de matriz cuando quiera calcular probabilidades de transición de un estado inicial a otro final):

$$\langle f|TT^\dagger |i\rangle = i(\langle f|T^\dagger |i\rangle - \langle f|T|i\rangle)$$

**Nota:** para procesos con simetría de inversión temporal del movimiento (QCD lo tiene): $\langle i|A|f\rangle = \langle f|A|i\rangle$, así que aunque el elemento de matriz del operador traspuesto conjugado cambie el orden de los kets, realmente podemos escribir todo como $T_{fi}$

- Introducimos ahora un conjunto completo de estados ortogonales en LHS entre los dos operadores:

$$\displaystyle\sum_j(2\pi)^4\delta(p_j-p_i)(2\pi)^4\delta^4(p_f-p_j) \langle f|T|j\rangle \langle j|T^\dagger |i\rangle = (2\pi)^4\delta^4(p_f-p_i)i(T_{fi}^*-T_{fi})$$

- De acuerdo a RHS, se debe conservar el cuadrimomento, de modo que para procesos físicos tenemos $p_i=p_f$ y por ende solo queda una delta

$$\boxed{2Im[T_{fi}]=(2\pi)^4\displaystyle\sum_jT_{fj}T_{ji}^*\delta^4(p_j-p_i)}$$

- Ahora, deseamos encontrar la condición de clausura en términos únicamente de los trimomentos: $I=\int d^3\vec qf(\vec q)|\vec q\rangle \langle \vec q|$

- Tenemos entonces que encontrar $f(\vec q)$, y lo hacemos comparando con lo que sabemos:


$$\langle p|p'\rangle = \langle p|I|p'\rangle = \langle E, \vec p|\left(\int d^3\vec qf(\vec q)|\vec q\rangle \langle \vec q|\right)|E', \vec p'\rangle$$

Ahora introduzco los kets y los actúo. Las energías van por su cuenta (**demostrar?**) y lógicamente sacan $\delta(E-E')$. En cuanto a los kets de trimomentos, también los hago actuar conforme a la normalización de Lorentz y tengo:

$$\langle p|p'\rangle = 4(2\pi)^6E_pE_p'\delta(E_p-E_p')\int d^3\vec q f(\vec q)\delta^3(\vec p-\vec q)\delta^3(\vec p'-\vec q)$$

Integrando sobre las q's (donde vea $\vec q$ pongo $\vec p$, por ejemplo), tengo finalmente:

$$\langle p|p'\rangle = 4(2\pi)^6E_pE_p'\delta(E_p-E_p') f(\vec p)\delta^3(\vec p'-\vec p)$$

Como tengo la delta en que los momentos han de ser idénticos, puedo suprimir la delta sobre las energías, porque estas dependen del módulo del momento. Por ende, ya admito que las dos energías que me salen son iguales y lo dejo como un cuadrado.

Imponiendo que sea igual a la cond. de normalización, saco qué tiene que valer $f(\vec p)$:

$$4(2\pi)^6E_p^2 f(\vec p)\delta^3(\vec p'-\vec p) = 2(2\pi)^3E_p\delta^3(\vec p'-\vec p)$$

Despejando, deduzco que:

$$f(\vec p)=\frac{1}{2(2\pi)^3E_p}$$

Y por ende la condición de clausura sobre estados de una partícula es:

$$I=\int\frac{d^3\vec p}{2(2\pi)^3E_p}$$

**Nota:** para sistemas de dos partículas, puedo se me ocurre un truco: puedo admitir que este estado de una sola partícula corresponde a la partícula efectiva del sistema de dos partículas, de modo que puedo invertir el razonamiento para pasar de vuelta al sistema de dos partículas, donde ahora tendré que la condición de clausura es:

Ahora quisiera integrar la delta, que va con una función de la variable de integración. Para dejarla en términos de la variable $k_n$, aplico:

$$\delta(g(k))=\displaystyle\sum_j\frac{\delta(k-k_j)}{|g'(k_j)|}$$

donde $k_j$ es la j-ésima raíz de $g(k)$. Como el argumento de la delta es una función con 2 raíces: $k_{n,0}=\pm \sqrt{\frac{s}{4}-m_n^2}$, podríamos pensar en incluir las dos en la suma de la delta, y esto sería lo correcto con la fórmula en la mano. $$k_{n,0}=+ \sqrt{\frac{s}{4}-m_n^2}$$

$$g'(k_n) = 2\frac{k_n}{\sqrt{k_n^2+m_n^2}}$$

donde la masa también lleva subíndice porque **partículas de diferentes canales tendrán en general distintas masas** ($m_\pi\neq m_K\neq m_\eta$)

$$\delta(2E_{k_n}-\sqrt s) = \frac{\delta(k_n-k_{n,0})}{2\sigma_n(s)}$$

## 3.4 Introducir desarrollo en ondas parciales

- Ahora que ya sabemos la expansión de la parte imaginaria de los elementos de matriz de $T$ en términos de productos de elementos de matriz, podemos introducir la expansión en ondas parciales y tratar de ver qué condición cumplen estas
- Recordamos:

$$T_{fi} = T(s, t, u)=16\pi\displaystyle\sum_{l=0}^\infty (2l+1)t_l^{fi}(s)P_l(\hat p\cdot \hat p')$$

- Calculemos primero la integral de ángulo sólido:

$$I=\int d\Omega_{\hat k_n}T_{fn}T_{ni}= \displaystyle\sum_{l,l'}(2l+1)(2l'+1)t_l^{fn}(s)(t_{l'}^{ni})^*(s)\int d\Omega_{\hat k_n}P_l(\hat p'\cdot \hat k_n)P_{l'}(\hat p\cdot \hat k_n)$$

### Teorema de integración de polinomios de Legendre

$$\int d\Omega_{\hat k}P_l(\hat a\cdot \hat k)P_{l'}(\hat b\cdot \hat k)=\frac{4\pi}{2l+1}\delta_{ll'}P_l(\hat a\cdot \hat b)$$

- Empleando este teorema, la integral es inmediata:

$$I=4\pi \displaystyle\sum_{l}(2l+1)t_l^{fn}(s)(t_{l}^{ni})^*(s)P_l(\hat p'\cdot \hat p)$$

Ahora introducimos en la expansión de $Im[T_{fi}]$ y comparamos:

$$16\pi \displaystyle\sum_l (2l+1)Im[t_l^{fi}(s)]P_l(\hat p'\cdot \hat p)=\frac{1}{64\pi^2}\displaystyle\sum_n \sigma_n(s)4\pi \displaystyle\sum_{l}(2l+1)t_l^{fn}(s)(t_{l}^{ni})^*(s)P_l(\hat p'\cdot \hat p)$$

Los factores numéricos se cancelan. 

**Observación:** como estamos trabajando a una energía dada $\sqrt s<\infty$, solo tendremos un número finito de estados intermedios de 2 partículas accesibles (i.e: no puedo crear un número infinito de canales porque llegará uno en el que las 2 partículas intermedias sean tan pesadas que no tenga energía para ello). Por ende, la suma en canales es **finita** y puedo intercambiar los dos sumatorios:

$$ \displaystyle\sum_l (2l+1)Im[t_l^{fi}(s)]P_l(\hat p'\cdot \hat p)= \displaystyle\sum_{l}(2l+1)P_l(\hat p'\cdot \hat p)\displaystyle\sum_n \sigma_n(s)t_l^{fn}(s)(t_{l}^{ni})^*(s)$$

Agrupando todo a un lado, escribo la expresión como un solo sumatorio que se anula:

$$ \displaystyle\sum_l (2l+1)\left(Im[t_l^{fi}(s)]-\displaystyle\sum_n \sigma_n(s)t_l^{fn}(s)(t_{l}^{ni})^*(s)\right)P_l(\hat p'\cdot \hat p)=0$$

Este sumatorio se anula para cualesquiera trimomentos de entrada y salida, de modo que lo que se anula ha de ser el término entre paréntesis. Por ende, encontramos que **la parte imaginaria de la onda parcial l-ésima para un proceso $a+b\to c+d$ es la suma de los productos de todas las ondas parciales l-ésimas de los procesos intermedios involucrados, pesados por el espacio de fases del proceso intermedio en cuestión**:

$$\boxed{Im[t_l^{fi}(s)] = \displaystyle\sum_n \sigma_n(s)t_l^{fn}(s)(t_l^{ni})^*(s)}$$

### Corolario: procesos elásticos

- Para un proceso elástico, donde las únicas partículas intermedias que puede haber coinciden con las de entrada (i.e: no creamos nuevas), tenemos un solo canal, $n=1$, de modo que, en este caso particular, la parte imaginaria de la onda parcial l-ésima del proceso se relaciona con su módulo al cuadrado (una especie de teorema óptico pero para cada onda parcial):

$$\boxed{Im[t_l^{fi}(s)] = \sigma(s)|t_l^{fi}(s)|^2}$$

### Objetivo 1: simetría de crossing

**Def.:** la simetría de crossing es el hecho de que, dado un proceso con unas partículas entrantes y otras salientes, si quitamos una de las entrantes y añadimos su antipartícula como saliente, la amplitud del proceso es la misma (siempre que el proceso siga siendo permitido). Matemáticamente, se ve como que la amplitud de un proceso con antipartículas es la continuación analítica a energías negativas de la amplitud del proceso con partículas ordinarias (hipótesis de Mandelstam)

**Nota:** dado el diagrama de un proceso, podemos "verlo" desde distintos canales. La elección de uno de ellos es arbitraria pero fija el resto. Por ejemplo, si tenemos A+B-->C+D y decimos que está en canal s, podemos "ver el diagrama desde arriba" y esto sería verlo en canal t, porque ahora las partículas entrates serían $A$ y $\bar C$, y las salientes serían $D$ y $\bar B$. Es decir, ver un proceso en un canal dado es lo mismo que hacer crossing con las partículas adecuadas. De este modo, lo que antes era la variable de Mandelstam $s$ pasa a ser $t$ y viceversa. $u$ queda intacto.

**Proposición:** en procesos a 2 piones, tenemos amplitudes con isospin bien definido, $F^I, \ \ I=0,1,2$

**Dem:**

(Siempre trabajamos en límite de isospin, i.e: $m_u=m_d$)

**Proposición:** en el límite de isospin, QCD es invariante bajo rotaciones en el espacio de isospin (simetría subyacente es $SU(2)$), de modo que el operador de scattering $S$ (equivalentemente, la parte de dispersión $T$) conmuta con los generadores del grupo:

$$[T, I_a]=0$$

**REF:** en "Invitation to QFT" se dice lo siguiente: "isospin remains only an approximaye symmetry of the strong force even after switching off both the EM and the weak interactions. This follows from the small but nonvanishing difference between the masses of the particles within an isospin multiplet. Isospin is nevertheless useful because the mass splitting is much smaller than the particle masses themselves and the symmetry breaking effects are small"

**Corolario:** por el teorema de Wigner-Eckart, entonces $S$ (o $T$) es escalar en este grupo de transformaciones, y por ende no es capaz de cambiar ninguno de los dos números cuánticos de isospin:

$$\langle I' I_3'|S|II_3\rangle = \langle I I_3;00|II_3'\rangle \langle I'||S||I\rangle \propto \delta_{II'}\delta_{I_3I_3'}$$

**Def.:** un pion vendrá dado en el espacio de isospin por un ket $|1 \ I_3\rangle$. Concretamente:

$$\begin{gather}|\pi^+\rangle \to |1 1\rangle \\ |\pi^0\rangle \to |1 0\rangle  \\ |\pi^-\rangle \to |1 -1\rangle  \end{gather}$$

Véase que esta no es la parte completa del vector de estado del pion, solo la parte de isospin. Por ejemplo, no ponemos la parte dinámica, que lleva el cuadrimomento, porque lo único que nos interesa es la parte de isospin. En procesos de dos piones, tenemos entonces que los isospines de los dos pion es se acoplan, y entonces el operador de scattering conmutará con los generadores del isospin total, claro, no de los piones individuales. Es entonces necesario trabajar en la base acoplada de isospin.

Consideremos el siguiente proceso: $\pi^+ \pi^- \longrightarrow \pi^0\pi^0$. Tenemos entonces que la amplitud del mismo (admitiendo que está en canal s) es, como se sabe de teoría de scattering:

$$A(s, t, u)=\langle \pi^0\pi^0|S|\pi^+\pi^-\rangle$$

Pero estos estados de la base desacoplada (i.e: $|\pi^+\pi^-\rangle \equiv |11;1-1|\rangle_D$) se pueden expresar asimismo en la base acoplada a través de los coeficientes de Clebsch-Gordan. Concretamente:

$$\begin{gather}|\pi^+\pi^-\rangle =\frac{1}{\sqrt 6}|20\rangle + \frac{1}{\sqrt 2}|10\rangle+\frac{1}{\sqrt 3}|00\rangle \\ |\pi^0\pi^0\rangle = \sqrt\frac{2}{3}|20\rangle-\frac{1}{\sqrt 3}|00\rangle\end{gather}$$

Sustituyendo en la amplitud del proceso:

$$A(s, t, u)=\frac{1}{3}F_2-\frac{1}{3}F_0$$

**Observación:** la amplitud en un canal dado de isospin es un objeto de por sí, no depende del proceso que estemos teniendo en cuenta. Este irá codificado en los coeficientes de Clebsh-Gordan que se empleen, pero fijado un isospin, la amplitud de un proceso en dicho canal ya es un elemento de matriz fijo

**Observación:** como en el límite de isospin el operador de scattering es invariante bajo transformaciones de esta naturaleza, podemos ignorar la tercera componente de isospin, que no contribuye información nueva. Por eso las amplitudes en canal de isospin $I$ solo dependen de cuál sea dicho isospin
