## 3.1 Lorentz invariance
- Einstein's principle of relativity: (in the absence of gravity) the laws of Physics are the same for all inertial frames of reference --> all (inertial) observers must agree on what the probability amplitude of a given process is: $\langle \alpha|S|\beta\rangle$ must be a scalar
- We choose the Lorentz-invariant normalization and demand $S$ to be a Lorentz scalar so that $T'=T$

## 3.2 Unitarity

### 3.2.1 Imposition
- If we begin with a given asymptotic state, we must end in another one --> $\displaystyle\sum_\beta |\langle \alpha |S |\beta\rangle|^2=\displaystyle\sum_\beta |\langle \alpha |S |\beta\rangle\langle \beta S^\dagger |\alpha\rangle = \langle \alpha| SS^\dagger |\alpha \rangle$ and this must be equal to the normalization

### 3.2.2 Consequences of unitarity

- I wouldn't mention optical theorem because it doesn't add anything to my project, maybe for forward dispersion relations if time allows

## 3.3 Causality

- Talk about the 3 forms of causality from Nussenszweig (classical, relativistic and micro)

## 3.4 Analytical structure of $T$

- How are causality and analyticity linked? --> Talk about Hilbert transforms and Tichmarsh theorem. Microcausality imposes an analytical structure in the scattering amplitude, although one can also view it classically with Tichmarsh theorem
- Let us promote $s$ as complex variable (and fix $t$). We can always recover the physical region by doing $T_{PHYS}(s, t)=\displaystyle\lim_{\varepsilon\to 0}T(s+i\varepsilon, t)$ (strictly, this comes from Tichmarsh theorem, so the validity of this prescription is to be checked later)
- **Bound states** --> poles in real axis
- **Resonances** --> poles in second Riemann sheet
- Scattering --> **branch cuts** --> one from threshold to infinity and one from -t to -infinity (crossing) (explain discontinuity in imaginary part)
- On real axis between -t and sth, T is real because there is no scattering (kinematically forbidden) --> this enables analytic continuation through **Schwartz reflection** principle --> we now have a function that is analytical everywhere except in poles and branch cuts
- **Crossing symmetry** is now explained naturally: it is the same amplitude, just in different regions of the complex plane!

## 3.5 Deriving dispersion relations

- Use **Cauchy theorem** to write $T(s, t)$
- Take **"comecocos"** contour (we assume no bound states because there are none in $\pi\pi$ scattering) --> take the arches to infinity --> now you have an **integral relation** between $T$ and $Im[T]$
- (Maybe talk about end-point singularities, pinching, etc.?)
- Using Sokovski-Plejenov formula (just mention) we recover $T_{PHYS}$ and now we take real part --> **dispersion relation** for $T$
- Talk about subtractions
    - Why do we need them? Plemelj formulae are only valid for square-integrable functions, and if they are not, the arches still contribute --> since the biggest $T$ can go is as $s\log^2(s)$ (**Froissart bound**) 2 subtractions are required to guarantee convergence, although in practice it can go slower and therefore only one subtraction may be in order (GKPY)
    - Derive how they are introduced in Nussenszweig

# 4. Narrowing down: pion scattering with Roy eqns.

- Objective of this section: explain the theoretical frame for $\pi\pi$ scattering and derive Roy equations (not in full!) --> this section unifies sections 2 and 3 to understand pion-pion scattering through dispersion relations

## 4.1 $\pi\pi$ scattering

- Imposition: isospin limit --> $[\vec I, S]=0$
- Therefore, we have $S$ is diagonal in coupled isospin basis (and indeed isospin-scalar from Wigner-Eckart theorem)
- Changing basis: ideal for 2->2 scattering is to work in coupled isospin basis (this is very basic $SU(2)$ irrep. addition)
- There are only 3 **independent** amplitudes from all pion scattering processes (the rest are recovered crossing to other channels): $T^I$
- How do these new well-defined isospin amplitudes relate to those of the uncoupled basis? --> Crossing matrices
- Link with 2: these relationships are true for every channel and in particular for well-defined isospin channels (quote thesis)

## 4.2 Deriving Roy equations
- You start with eqn. A8
- Re-express everything in terms of ds' --> using crossing matrix Csu
- 2 subtractions ==> we need 2 external parameters --> evaluate T at selected points in s, t (and u, on-shell) --> threshold, since we can then relate it to scattering lengths, which are directly measurable --> the 2 parameters are scattering length of S wave in I=0,2 channels
- We arrive at eqn. A.23
- We now project in partial waves and get to the Roy eqns.
- Briefly: from partial waves one can extract phase shifts (for instance, they tell us when there is a resonance) and inelastic factors (they tell us when scattering ceases to be elastic and we "lose probability" to other channels we are not considering)

## 4.3 Going deeper into the Roy equations
