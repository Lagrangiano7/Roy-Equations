# Oscilador armónico forzado

- Sea un oscilador armónico forzado por una fuerza externa $F(t)$ y en general arrastrado con constante de damping $\gamma>0$. Su ecuación del movimiento es:

$$\ddot x+2\gamma\dot x+\omega_0^2x=f(t)$$

## Funciones de Green

- La solución homogénea la conozco (la amplitud oscila mientras decae exponencialmente)
- Me centro en la solución particular

**Teorema:** dado un operador diferencial lineal $L(x)$ que actúe en general sobre alguna distribución $u(x)$ en cierto subespacio de $\mathbb R^n$, una solución particular de la ecuación será:

$$u(x)=(g*f)[t] = \int_{\mathbb R} g(x, s)f(s)ds$$

**Dem.:**

Sea una función $g(x, s)$ tal que dé la fuente:

$$Lg(x, s)=\delta(x-s)$$

Actuamos la fuente del problema original por ambos lados e integramos en $s$. Vemos que entonces obtenemos $f(x)$ en el lado derecho, de modo que el lado izquierdo debe ser igual a $Lu(x)$, pero como $L$ es lineal por hipótesis, podemos sacarlo de la integral, de modo que la integral debe coincidir con la solución $u(x)$:

$$\int Lg(x, s)f(s)ds = \int f(s)\delta(x-s)ds = f(x)$$

$$\therefore Lu(x)=\int Lg(x, s)f(s)ds =L\int g(x, s)f(s)ds \Longrightarrow u(x)=\int g(x, s)f(s)ds_\blacksquare$$

**Observación:** encontrar la función de Green correspondiente es entonces equivalente a resolver el problema.

**Proposición:** cuando el operador $L(x)$ es invariante bajo traslaciones en su parámetro, entonces el operador de Green va como $g(x, s)=g(x-s)$

## Fin de funciones de Green

- Entonces, para hallar una solución particular de este problema, basta con hacer:

$$x(t)=\int_{-\infty}^\infty g(t-t')f(t')dt'$$

- Encontremos ahora cuál es la función de Green para resolver el problema. Para ello, descompongamos nuestras funciones en la base de Fourier:

$$x(t)=\frac{1}{2\pi}\int_\mathbb R X(\omega)e^{-i\omega t}d\omega; \ \ f(t)=\frac{1}{2\pi}\int_\mathbb R F(\omega)e^{-i\omega t}d\omega $$

- Ahora introducimos estas funciones en la ecuación, obteniendo una relación para los coeficientes (como conocemos el input, se entiende que $F(\omega)$ también es conocida, porque podemos hallarla a partir del input como una IFT. Tenemos que:

$$X(\omega)=-\frac{F(\omega)}{(\omega-\omega_1)(\omega-\omega_2)}=G(\omega)F(\omega)$$

donde $\omega_{1, 2}=\pm \sqrt{\omega_0^2-\gamma^2}-i\gamma$

Ya está, porque como ahora podemos volver a integrar:

$$x(t)=\frac{1}{2\pi}\int_\mathbb R X(\omega)e^{-i\omega t}d\omega = -\frac{1}{2\pi}\int_\mathbb R\frac{F(\omega)}{(\omega-\omega_1)(\omega-\omega_2)}e^{-i\omega t}d\omega$$

Como $F(\omega)=\int_\mathbb Rf(t')e^{i\omega t'}dt'$ podemos reintroducirlo:

$$x(t)=\frac{1}{2\pi}\int_\mathbb Rf(t')dt'\int_\mathbb RG(\omega)e^{-i\omega(t-t')}d\omega$$



Si definimos la función de Green:

$$g(t-t')=\int_\mathbb R G(\omega)e^{-i\omega(t-t')}d\omega$$

Entonces ya tenemos el problema en la forma del teorema:

$$\boxed{x(t)=\int_\mathbb R g(t-t')f(t')dt'}$$

y el problema ha quedado resuelto.

**Observación:** comprobamos que la función de Green sale con la diferencia entre parámetros porque el operador diferencial es invariante bajo traslaciones en $t$

**Importante:** esperaríamos que lo que le pase a la partícula solo venga influenciado por lo que ha sucedido en el pasado, no por lo que pase en el futuro. De este modo, esperaríamos que el límite de integración superior fuera hasta $t$. Vemos que $g(t-t')=0; \ \forall t <t'$

Definamos por comodidad la función de Green dependiente del intervalo $\tau=t-t'$:

$$g(\tau)=\int_\mathbb R G(\omega)e^{-i\omega\tau}d\omega$$

Como los dos polos tienen parte imaginaria negativa, ambos están en $I_-$, de modo que si queremos integrar para $\tau<0$, podemos hacerlo cerrando el contorno por arriba ($I_+$). Como la función $G(\omega)$ es analítica en $I_+$, la integral es 0 por el teorema de Cauchy. Llegamos entonces al importante resultado:

$$\boxed{g(\tau)=0; \ \ \forall \tau<0}$$

Como intuíamos, el sistema solo se ve afectado por lo que le ha sucedido en el pasado, no por lo que vaya a suceder en el futuro. Parece haber una estrecha relación entre la analiticidad de una función y la causalidad en el sistema
